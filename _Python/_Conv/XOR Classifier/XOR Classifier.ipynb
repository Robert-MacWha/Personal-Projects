```
import tensorflow.keras as keras    # Keras is a library designed to allow users to more easily write machein learning algorythms by simplyfying required code
import matplotlib.pyplot as plt     # MatPlotLib is a library used to display images and graphs
```

```
# Defining dataset used to train the model
Xs = [[0, 0], [0, 1], [1, 0], [1, 1]]
ys = [[0], [1], [1], [0]]
```

```
# Create a sequential model and populate it with two layers
model = keras.models.Sequential()
model.add(keras.layers.Dense(input_dim=(2), units=8, activation="relu"))
model.add(keras.layers.Dense(units=1, activation="sigmoid"))

print(model.summary())

# Compile the model with the adam optimizer and the mean squared error loss function
model.compile(optimizer="adam", loss="mse")
```

```
# Train the model on the pre-defined dataset.  In this case the model will be trained for 1000 epochs
model.fit(Xs, ys, epochs=1000)
```
